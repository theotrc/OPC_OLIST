# Merge

### Monetary

```python
df_pymt_value = df_orders.merge(df_pymts[["order_id", "payment_value"]], on="order_id", how="left")[["payment_value", "order_id", "customer_id"]]
df_total_order = df_pymt_value.groupby(by="order_id").sum().reset_index()
df_total_order
```

```python
df_monetary = df_customers.merge(df_total_order[["customer_id", "payment_value"]], on="customer_id", how="left")[["customer_unique_id", "payment_value"]].rename(columns={"payment_value":"Monetary"})
df_monetary = df_monetary.groupby(by="customer_unique_id").sum().reset_index()
```

### recency

```python
df_orders["order_purchase_timestamp"] = pd.to_datetime(df_orders["order_purchase_timestamp"], format="%Y-%m-%d %H:%M:%S")
```

```python
df_orders
```

```python
df_recency = df_orders[["order_id", "customer_id", "order_purchase_timestamp"]]
df_recency = df_recency.merge(df_customers[["customer_id", "customer_unique_id"]], how="left", on="customer_id")
df_recency = df_recency.groupby(by = "customer_unique_id").max().reset_index()[["customer_unique_id", "order_purchase_timestamp"]].rename(columns={"order_purchase_timestamp":"Recency"})
```

```python
df_recency
```

### Frequency

```python
df_frequency = df_customers[["customer_id", "customer_unique_id"]].groupby(by="customer_unique_id").count().rename(columns={"customer_id":"Frequency"}).reset_index()
df_frequency
```

### RFM

```python
df_rfm = df_recency.merge(df_monetary, how="left", on="customer_unique_id").merge(df_frequency, how="left", on="customer_unique_id")
ref_date = df_rfm["Recency"].max()
```

```python
df_rfm["Recency"] = df_rfm["Recency"].apply(lambda x: (ref_date - x).days)
```

```python
df_rfm
```

```python
plt.hist(df_rfm["Monetary"], bins=50, edgecolor="k")
plt.show()
```

```python
plt.boxplot(df_rfm["Monetary"])
plt.show()
```

```python
df_rfm.to_csv("data/RFM.csv", index=False)
```

## Reviews => customers

```python
df_review_score_mean = df_reviews[["order_id", "review_score"]].groupby(by="order_id").mean().reset_index()
```

```python
df_review_score_mean
```

```python
df_review_score = df_orders.merge(df_review_score_mean[["order_id", "review_score"]], how="left", on="order_id")[[ "review_score", "customer_id"]]
df_review_score = df_customers.merge(df_review_score, how="left", on="customer_id")[["customer_unique_id", "review_score"]].groupby(by="customer_unique_id").mean().reset_index()
df_review_score
```

```python
df_rfm = df_rfm.merge(df_review_score, how="left", on="customer_unique_id").rename(columns={"review_score":"mean_review_score"})
df_rfm = df_rfm.merge(df_customers[["customer_unique_id", "customer_state"]], how="left", on="customer_unique_id").drop_duplicates()
```

## Calculate the distance

### adding geolocation data to customer

```python
df_lat_lng = df_geoloc[["geolocation_lat", "geolocation_lng", "geolocation_city"]].groupby(by=["geolocation_city"]).first().reset_index()
df_customers_pos = df_customers.rename(columns={"customer_city":"geolocation_city"}).merge(df_lat_lng[["geolocation_lat", "geolocation_lng", "geolocation_city"]], on="geolocation_city", how="left")[["customer_id","customer_unique_id", "geolocation_lat", "geolocation_lng"]]
df_customers_pos
```

```python
df_sellers_pos = df_sellers.rename(columns={"seller_city":"geolocation_city"}).merge(df_lat_lng[["geolocation_lat", "geolocation_lng", "geolocation_city"]], on="geolocation_city", how="left")[["geolocation_city","seller_id", "geolocation_lat", "geolocation_lng"]]
df_sellers_pos = df_sellers_pos.dropna()
df_sellers_pos
```

```python
df_items_pos = df_order_items.join(df_sellers_pos.set_index("seller_id"), on="seller_id")
df_items_pos = df_items_pos[["order_id", "geolocation_lat", "geolocation_lng"]].copy()
```

```python
df_items_pos = df_items_pos.groupby("order_id").agg("mean")
```

```python
df_orders_pos = df_orders[["order_id", "customer_id"]].join(df_items_pos, on="order_id")
```

```python
df_orders_pos
```

Connect a customer to his order's positions


```python
df_customers_pos
```

```python
df_customers_pos  = df_customers_pos.join(df_orders_pos.set_index("customer_id"), on="customer_id", rsuffix="_order").drop("order_id", axis=1)
```

```python
df_customers_pos
```

```python
df_customers_pos["distance"] = np.sqrt(
  (df_customers_pos["geolocation_lat"] - df_customers_pos["geolocation_lat_order"])**2 +
  (df_customers_pos["geolocation_lng"] - df_customers_pos["geolocation_lng_order"])**2
  )
```

```python
df_customers_pos.drop(["geolocation_lat", "geolocation_lng", "geolocation_lat_order", "geolocation_lng_order"], axis=1, inplace=True)
```

```python
df_customers_pos
```

```python
df_customers_pos = df_customers_pos.groupby("customer_unique_id").agg({"distance": "mean"}).reset_index()
df_customers_pos
```

```python
df_rfm = df_rfm.merge(df_customers_pos, how="left", on="customer_unique_id").rename(columns={"distance":"mean_distance"})
```
